```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(8, 4))

gs = gridspec.GridSpec(1, 2, width_ratios=[3, 5])  # 3 units and 5 units

ax1 = plt.subplot(gs[0])
ax2 = plt.subplot(gs[1])

ax1.text(0.5, 0.5, "3 units", ha='center', va='center', fontsize=15)
ax2.text(0.5, 0.5, "5 units", ha='center', va='center', fontsize=15)

plt.show()
```
